```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

UsedCar_data = pd.read_csv("Usedcar.csv") # read the data from dataset

UsedCar_data.isna().sum() # checking if there is any missing values in dataset
```

```python
def bar_plot(Categorical_Col):
    #get variable
    colvar = UsedCar_data[Categorical_Col]
    #count number of categorical variable(value/sample)
    colvarValue = colvar.value_counts()
    
    #visualize
    plt.figure(figsize=(9,9))
    plt.bar(colvarValue.index, colvarValue)
    plt.xticks(colvarValue.index, colvarValue.index.values)
    plt.ylabel("count")
    plt.title(Categorical_Col)
    plt.show()
    print("{}: \n {}".format(Categorical_Col,colvarValue))
```

```python
category1 = ["brand","model","transmission","fuelType"]
for c in category1:
    bar_plot(c)
```

```python
def plot_hist(Numerical_var):
    plt.figure(figsize=(9,3))
    plt.hist(UsedCar_data[Numerical_var], bins=100)
    plt.xlabel(Numerical_var)
    plt.ylabel("Frequency")
    plt.title(f" {Numerical_var} distribution with hist")
    plt.show()
```

```python
numeric_var = ["year","mileage","tax","mpg","engineSize"]
for i in numeric_var:
    plot_hist(i)
```
